### 2. Data Preparation for modelling

```python
inp2 = inp1.copy()
```

```python
#Before dropping date, let us introduce a days_old variable which indicates how old is the business
inp2['days_old'] = (pd.to_datetime(inp2['dteday'],format= '%d-%m-%Y') - pd.to_datetime('01-01-2018',format= '%d-%m-%Y')).dt.days
```

```python
inp2.drop(["dteday","instant",'casual','registered'] , axis = 1 , inplace = True)
```

#### 2.1 Dummy Variables

```python
inp2 = pd.get_dummies(data=inp2,columns=["season","mnth","weekday"],drop_first=True)
inp2 = pd.get_dummies(data=inp2,columns=["weathersit"])
# We dont drop any column in weathersit because already out of 4 only 3 are present, 
# year is already present only in 1 column
inp2.head()
```

```python
inp2.info()
```

#### 2.2 Test - Train Split

```python
from sklearn.model_selection import train_test_split
```

```python

#y to contain only target variable
y=inp2.pop('cnt')

#X is all remainign variable also our independent variables
X=inp2

#Train Test split with 70:30 ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=100)
```

```python
# Checking shape and size for train and test
print(X_train.shape)
print(X_test.shape)
```

#### 2.3 Feature Scaling continuous variables

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Let us scale continuous variables
num_vars = ['temp','atemp','hum','windspeed','days_old']

#Use Normalized scaler to scale
scaler = MinMaxScaler()

#Fit and transform training set only
X_train[num_vars] = scaler.fit_transform(X_train[num_vars])

#Inspect stats fro Training set after scaling
X_train.describe()
```

### 3A. Training the model

```python
# There are 31 Variables  we will use RFE to bring the number of variables to around 15 and then perform manual elimination
```

#### 3A.1 RFE for building a model

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

# running RFE
rfe = RFE(lm, 15)             
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
X_train.columns[rfe.support_]
```

#### 3A.2 Function to build a model using statsmodel api

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def build_model(cols):
    X_train_sm = sm.add_constant(X_train[cols])
    lm = sm.OLS(y_train, X_train_sm).fit()
    print(lm.summary())
    return lm
```

#### 3A.3 Function to calculate VIFs and print them

```python
def get_vif(cols):
    df1 = X_train[cols]
    vif = pd.DataFrame()
    vif['Features'] = df1.columns
    vif['VIF'] = [variance_inflation_factor(df1.values, i) for i in range(df1.shape[1])]
    vif['VIF'] = round(vif['VIF'],2)
    print(vif.sort_values(by='VIF',ascending=False))
```

### 3B. Training the model - Building the model

#### Model 1

```python
cols = list(X_train.columns[rfe.support_])
build_model(cols)
get_vif(cols)
```

```python
cols.remove('days_old')
build_model(cols)
get_vif(cols)
```

#### Model 5 fits our parameters P values of all variables is less than 5%; R sq is around 78%; Prob(F Statistic) is about zero ; VIF is less than 5 so build the model with above variables

```python
#Build a model with above columns
def build_model_sk(X,y):
    lr1 = LinearRegression()
    lr1.fit(X,y)
    return lr1

#Build a model with above columns
lr = build_model_sk(X_train[cols],y_train)
print(lr.intercept_,lr.coef_)
```

### 4. Residual Analysis

```python
y_train_pred = lr.predict(X_train[cols])
#Plot a histogram of the error terms
def plot_res_dist(act, pred):
    sns.distplot(act-pred)
    plt.title('Error Terms')
    plt.xlabel('Errors')
```

```python
plot_res_dist(y_train, y_train_pred)
```

#### Errors are normally distribured here with mean 0. So everything seems to be fine

#### 5.2 Making predictions on test set

```python
#Making Predictions
#Applying the scaling on the test sets
X_test[num_vars] = scaler.transform(X_test[num_vars])
#Predict the values for test data
y_pred = lr.predict(X_test[cols])
```

```python
#Model Evaluation
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16)                          # Y-label
```

#### 5.3 Find out the R squared value between test and predicted test data sets.

```python
# Find out the R squared value between test and predicted test data sets.  
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
## R2 is 0.79 which is quite close to training set r2 value
```

#### 5.4 Plotting error terms.

```python
# Error terms
def plot_err_terms(act,pred):
    c = [i for i in range(1,len(act)+1,1)]
    fig = plt.figure(figsize=(15,8))
    plt.plot(c,act-pred, color="purple",marker = 'o', linewidth=6, linestyle="")
    fig.suptitle('Error Terms', fontsize=30)              # Plot heading 
    plt.xlabel('Index', fontsize=25)                      # X-label
    plt.ylabel('Counts - Predicted Counts', fontsize=25) 
```

```python
#Plot error terms for test data
plot_err_terms(y_test,y_pred)
```
